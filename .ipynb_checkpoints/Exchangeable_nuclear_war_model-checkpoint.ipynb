```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

import jax.numpy as jnp
from jax import random, vmap

import numpyro
import numpyro.distributions as dist
from numpyro.infer import MCMC, NUTS, Predictive
sns.set()
```

```python
# If we want to use a beta distribution for the prior, it's helpful to see what different parameters look like

xs = jnp.linspace(0,0.1,500)
fig,ax = plt.subplots(figsize=(14,6))
tests = [[0.5,10],[5,100],[50,1000]]
for a,b in tests:
    m = dist.Beta(a,b).mean
    v = np.sqrt(dist.Beta(a,b).variance)
    ax.plot(xs,dist.Beta(a,b).cdf(xs),label=f"a={a},b={b}, mean {m:.2f}, std {v:.3f}")
    
plt.legend()
```

```python
def model(a,b,iid):
    if iid:
        prior_rate=iid
    else:
        # forecasted distribution over annual probability of nuclear war
        prior_rate = numpyro.sample('prior probability',dist.Beta(a,b))
    with numpyro.plate('w',100):
        war = numpyro.sample('war',dist.Bernoulli(prior_rate),infer={'enumerate':'parallel'})
```

```python
def run_samples_and_predict(model,
                a=0.5,
                b=10,
                iid=False,
                num_warmup = 2000, 
                num_samples = 2000):
    # Start from this source of randomness. We will split keys for subsequent operations.
    rng_key = random.PRNGKey(0)
    rng_key, rng_key_ = random.split(rng_key)
     # Run NUTS.
    kernel = NUTS(model)
    mcmc = MCMC(kernel, num_warmup=num_warmup, num_samples=num_samples)
    mcmc.run(rng_key_, a,b,iid)
    samples = mcmc.get_samples()
    predict = Predictive(model, samples, infer_discrete=True, num_samples=num_samples)
    predict_samples = predict(random.PRNGKey(1), a, b, iid)
    return predict_samples
```

```python
# Predict war over the next 100 years with low confidence in rate
_, predict1 = run_samples_and_predict(model)
# Predict war over the next 100 years with higher confidence in rate
_, predict2 = run_samples_and_predict(model,a=5,b=100)
```

```python
def has_war_occurred(predict):
    hw = np.zeros(100)
    for i in range(predict['war'].shape[1]):
        hw[i] = (predict['war'][:,:i].sum(1)>0).sum()/predict['war'].shape[0]
    return hw
```
